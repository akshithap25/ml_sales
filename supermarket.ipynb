```python
sns.set(style="darkgrid")
genderCount  = sns.countplot(x="Gender", data =data).set_title("Gender_Count")
```

```python
data['Time'] = pd.to_datetime(data['Time'])
data['Hour'] = (data['Time']).dt.hour
genderCount  = sns.lineplot(x="Hour",  y = 'Quantity',data =data).set_title("Product Sales per Hour")
```

```python
sns.boxplot(x="Branch", y = "Rating" ,data =data).set_title("Ratings by Branch")
```

```python
sns.boxenplot(y = 'Product line', x = 'Quantity', data=data )
#From the visual, Health and Beauty,Electronic accessories, Home and lifestyle, Sports and travel have a better average quantity sales that food and beverages as well as Fashion accessories.
```

```python
sns.boxenplot(y = 'Product line', x = 'Rating', data=data )
#Food and Beverages have the highest average rating while sports and travel the lowest
```

```python
sns.countplot(y = 'Product line', data=data, order = data['Product line'].value_counts().index )
#image shows the top product line item type sold in the given dataset. Fashion Accessories is the highest while Health and beauty is the lowest
```

```python
sns.countplot(x="Customer type", data =data).set_title("Customer Type")
```

```python
sns.countplot(x="Payment", data =data).set_title("Payment Channel")
#Most of the customers pay through the Ewallet and Cash Payment while under 40 percent of them pay with their credit card.
#We would also like to see this payment type distribution across all the branches
```

```python
sns.countplot(x="Payment", hue = "Branch", data =data).set_title("Payment Channel by Branch")
```
